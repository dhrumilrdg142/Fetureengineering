```python
db.columns = db.columns.str.strip()

# Handle missing values (fill or drop)
db.fillna('Unknown', inplace=True)

# -----------------------------------------------------
# 3️⃣ Feature Engineering
# -----------------------------------------------------

# 🧩 (a) Extract Stipend range
# Convert '₹ 10,000 - 15,000 /month' → min_stipend=10000, max_stipend=15000, avg_stipend=12500
def parse_stipend(stipend):
    try:
        stipend = stipend.replace('₹', '').replace('/month', '').replace(',', '').strip()
        if '-' in stipend:
            parts = stipend.split('-')
            min_s = int(parts[0])
            max_s = int(parts[1])
            return pd.Series([min_s, max_s, (min_s + max_s)//2])
        else:
            value = int(stipend)
            return pd.Series([value, value, value])
    except:
        return pd.Series([0, 0, 0])

db[['min_stipend', 'max_stipend', 'avg_stipend']] = db['Stipend'].apply(parse_stipend)

```

```python
db['duration_months'] = db['Duration'].str.extract(r'(\d+)').fillna(0).astype(int)

# -----------------------------------------------------
# 🧩 (c) Extract Apply-by Date features
# Convert "16 Oct' 25" → datetime
db['Apply by Date'] = pd.to_datetime(db['Apply by Date'].str.replace("'", " 20"), format='%d %b %Y', errors='coerce')

# Extract date components
db['apply_year'] = db['Apply by Date'].dt.year
db['apply_month'] = db['Apply by Date'].dt.month
db['apply_day'] = db['Apply by Date'].dt.day

# -----------------------------------------------------
# 🧩 (d) Count number of skills & perks
db['num_skills'] = db['Skills'].apply(lambda x: len(x.split(',')))
db['num_perks'] = db['Perks'].apply(lambda x: len(x.split(',')))
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ['profile', 'company', 'Location', 'Offer', 'Education']:
    le = LabelEncoder()
    db[col] = le.fit_transform(db[col])

# -----------------------------------------------------
# 🧩 (f) Handle Start Date
# If “Immediately” → 0, otherwise convert to numeric months ahead
db['Start Date'] = db['Start Date'].apply(lambda x: 0 if 'immediate' in x.lower() else 1)

# -----------------------------------------------------
# 4️⃣ Drop unused columns (like raw text)
db.drop(columns=['Stipend', 'Duration', 'Apply by Date', 'Skills', 'Perks'], inplace=True)
```
